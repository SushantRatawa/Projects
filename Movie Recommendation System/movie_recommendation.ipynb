NLP
1. remove stop words and punctuators
2. tokenization - splitting the words
3. using lemmatization

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
  #lowercase
  text = str(text).lower()
  #remove punctuators
  text = re.sub(r'[^a-zA-Z\s]','',text)
  #tokenization
  words = text.split()
  #removing stopwords
  words = [word for word in words if word not in stop_words]
  #lemmatizing
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
movie_df['tags'] = movie_df['tags'].apply(preprocess_text)
```

```python
movie_df = movie_df.reset_index(drop = True)
```

```python
#converts every char to lower and remove if any leading or trailing space is there
movie_df['title_clean'] = movie_df['title'].str.lower().str.strip()

indices =pd.Series(movie_df.index,index = movie_df['title_clean']).drop_duplicates()
indices
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words = 'english')
```

```python
tfidf_matrix = tfidf.fit_transform(movie_df['tags'])
```

```python
tfidf_matrix
```

```python
print(tfidf_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend(title,n=10):
  title = title.lower().strip()

  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  sim_score = cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
  similar_idx = sim_score.argsort()[::-1][1:n+1]
  return movie_df['title'].iloc[similar_idx]
```

```python
recommend("thor")
```
